```python


#Delete after running once



first_sim = rebound.Simulation()
first_sim.add(m=1) 
first_sim.add(m=1*mass_ratio, P = 1, e = 0.00)
first_sim.add(m=1*mass_ratio, P = 1.3, e = 0.00)
first_sim.add(m=1*mass_ratio, P = 1.5, e = 0.00)

first_prob = fmodel.predict_stable(first_sim)



print("The first prob is: " + str(first_prob))




second_sim = rebound.Simulation()
second_sim.add(m=0.1) 
second_sim.add(m=1*mass_ratio, P = 1, e = 0.00)
second_sim.add(m=1*mass_ratio, P = 1.3, e = 0.00)
second_sim.add(m=1*mass_ratio, P = 1.5, e = 0.00)

second_prob = fmodel.predict_stable(second_sim)



print("The second prob is: " + str(second_prob))



third_sim = rebound.Simulation()
third_sim.add(m=100) 
third_sim.add(m=1*mass_ratio, P = 1, e = 0.00)
third_sim.add(m=1*mass_ratio, P = 1.3, e = 0.00)
third_sim.add(m=1*mass_ratio, P = 1.5, e = 0.00)

third_prob = fmodel.predict_stable(third_sim)



print("The third prob is: " + str(third_prob))

```
